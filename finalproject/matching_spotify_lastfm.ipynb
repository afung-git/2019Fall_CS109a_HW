```python
import pandas as pd
import sqlite3
import numpy as np
```

```python
## connect to sql database
meta="track_metadata.db"
conn_meta  = sqlite3.connect(meta)

## data query
sql  = """SELECT track_id, title, artist_name FROM songs """
meta_data = np.array(conn_meta.execute(sql).fetchall())
```

```python
metadata_pd=pd.DataFrame(meta_data)
```

```python
metadata_pd=metadata_pd.rename(columns={0:'lastfm_track_id',1:'track_name',2:'artist_name'})
metadata_pd.head()
```

```python
lastfm_with_tags=pd.read_csv('lastfm_with_tags.csv')
```

```python
lastfm_with_tags.head()
```

```python
lastfm_names_tags=pd.merge(lastfm_with_tags,metadata_pd,on='lastfm_track_id').drop(['Unnamed: 0'],axis=1)
lastfm_names_tags.head()
```

```python
lastfm_names_tags.shape
```

```python
songs_info=pd.read_csv('tracks_features_info.csv').drop(['Unnamed: 0'],axis=1)
```

```python
songs_info.head()
```

```python
songs_info.shape
```

```python
combined=pd.merge(lastfm_names_tags, songs_info, on=['track_name','artist_name'],how='outer')
```
