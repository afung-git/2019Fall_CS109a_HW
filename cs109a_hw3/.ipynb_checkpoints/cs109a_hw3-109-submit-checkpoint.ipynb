Obviously the residuals are not normally distributed, and the residuals are in a pattern with predicted y. Plus, the residuals are not scattered above and below 0 randomly. It means the assumptions of linearity, constant variance and normality don't hold.

**4.2 Write a function to standardize...**

```python
# your code here
def standard(df,df_reference,columns=None):
    if columns==None:
        columns=list(df.columns)
    df_scaled=df.copy()
    for column in columns:
        mu=df_reference[column].mean()
        sigma=df_reference[column].std()
        df_scaled[column]=(df_scaled[column]-mu)/sigma
    return df_scaled

binary_columns = [ 'holiday', 'workingday','Feb', 'Mar', 'Apr',
       'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec', 'spring',
       'summer', 'fall', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat',
       'Cloudy', 'Snow', 'Storm','year']

practice_X_train=BSS_train[X_train][BSS_train[X_train].columns.difference(binary_columns)]
practice_X_train=practice_X_train.drop(columns=['atemp'])
practice_X_train.head()
practice_X_train_scaled=standard(practice_X_train,practice_X_train)
practice_X_train_scaled.describe()
```

The mean of all columns can be condisered 0, and the standard error of all columns are 1. All columns are standardized.

**4.3 Write code to augment your a dataset with higher-order features for...**

```python
# your code here
def addpoly(df,degrees:list,columns=None):
    df_addpoly=df.copy()
    if columns == None:
        columns=list(df.columns)
    for degree in degrees:
        if degree>1:
            for column in columns:
                df_addpoly[column+'^'+str(degree)]=df[column]**degree
    return df_addpoly

practice_X_train_poly=addpoly(practice_X_train_scaled,[2,3])
print(practice_X_train_poly.columns)
practice_X_train_poly.head()
```

**4.4 Write code to add interaction terms to the model...**

```python
# your code here
def addinteraction(df_new,df_source,list_cont,list_binary):
    df_addinter=df_new.copy()
    for cont in list_cont:
        for binary in list_binary:
            df_addinter[cont+"*"+binary]=df_new[cont]*df_source[binary]
    return df_addinter
    
list_cont=list(practice_X_train_scaled.columns)[1:]
list_binary=list(BSS_train_drop.columns)[9:26]
practice_X_train_poly_addinter=addinteraction(practice_X_train_poly,BSS_train_drop,list_cont,list_binary)
print(practice_X_train_poly_addinter.columns)
practice_X_train_poly_addinter.head()
```

**4.5 Combine all your code so far into a function named...**

```python
def get_design_mats(train_df, val_df,  degree, 
                    columns_forpoly=['temp', 'hum','windspeed', 'hour'],
                    target_col='counts', 
                    bad_columns=['workingday','atemp','spring','summer','fall','casual','registered'],
                    list_cont=['hum', 'temp', 'windspeed'],
                    list_binary=['Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 
                                 'Oct', 'Nov', 'Dec', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']):
    """
    :param train_df: The training data
    :param val_df: Your test/validation data
    :param columns_forpoly: List containing the column name for which we want to create 
                            polynomial terms
    :param target_col: The response variable column
    :param bad_columns: The columns you want to drop
    """    
    # add code here 
    # 1. drop the bad columns 
    train_df=train_df.drop(columns=bad_columns)
    val_df=val_df.drop(columns=bad_columns)
    # 2. standardize continuous columns
    reference=train_df.copy()
    train_df=standard(train_df,reference,columns=columns_forpoly)
    val_df=standard(val_df,reference,columns=columns_forpoly)
    # 3. add poly terms
    train_df=addpoly(train_df,degrees=list(range(1,degree+1)),columns=columns_forpoly)
    val_df=addpoly(val_df,degrees=list(range(1,degree+1)),columns=columns_forpoly)
    # 4. add interaction terms
    train_df=addinteraction(train_df,train_df,list_cont,list_binary)
    val_df=addinteraction(val_df,val_df,list_cont,list_binary)
    # 5. final steps
    x_train=train_df.drop(columns=['counts'])
    y_train=train_df['counts']
    x_val=val_df.drop(columns=['counts'])
    y_val=val_df['counts']
    
    return x_train,y_train, x_val,y_val
```

**5.1 Let's try regularizing our models via...** 

```python
# your code here
from sklearn.linear_model import Ridge
BSS_train=pd.read_csv('data/BSS_train.csv').drop(columns=['dteday'])
BSS_test=pd.read_csv('data/BSS_test.csv').drop(columns=['dteday'])
reg_train,reg_val=train_test_split(BSS_train,test_size=0.2,random_state=42)
reg_x_train_ls=[]
reg_y_train_ls=[]
reg_x_val_ls=[]
reg_y_val_ls=[]
reg_mod_ls=[]
reg_score_val_ls=[]
reg_score_train_ls=[]
ld=[.01,.05,.1,.5,1,5,10,50,100]
for i in list(range(1,9)):
    x_train_i,y_train_i, x_val_i,y_val_i=get_design_mats(reg_train,reg_val,i,
                                                         bad_columns=['workingday','spring','summer','fall','casual','registered'])
    x_train_ls.append(x_train_i)
    y_train_ls.append(y_train_i)
    x_val_ls.append(x_val_i)
    y_val_ls.append(y_val_i)
    mod_i=[]
    score_val=[]
    score_train=[]
    for each in ld:
        ridge_model = Ridge(alpha=each).fit(x_train_i, y_train_i)
        mod_i.append(ridge_model)
        score_val.append(ridge_model.score(x_val_i,y_val_i))
        score_train.append(ridge_model.score(x_train_i,y_train_i))    
    reg_score_train_ls.append(score_train)
    reg_score_val_ls.append(score_val)
    reg_x_train_ls.append(x_train_ls)
    reg_x_val_ls.append(x_val_ls)
    reg_y_train_ls.append(y_train_ls)
    reg_y_val_ls.append(y_val_ls)
    reg_mod_ls.append(mod_i)

output=pd.DataFrame(np.array(reg_score_val_ls))
output.columns=ld
output.index=range(1,9)
output
```
